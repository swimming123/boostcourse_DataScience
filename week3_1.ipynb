```python
age_code = {
 5: '20~24세',
 6: '25~29세',
 7: '30~34세',
 8: '35~39세',
 9: '40~44세',
 10: '45~49세',
 11: '50~54세',
 12: '55~59세',
 13: '60~64세',
 14: '65~69세',
 15: '70~74세',
 16: '75~79세',
 17: '80~84세',
 18: '85세+'}
```

```python
df['허리둘레'].head()
```

```python
df['연령대코드(5세단위)'].head()
```

```python
#df_group 으로 연령대코드(5세단위) 기준 허리둘레 describe()
df_group = df.groupby(['연령대코드(5세단위)'])['허리둘레'].describe()
df_group
```

```python
#df_group 타입 확인
type(df_group)
```

```python
#age_code 타입 시리즈로 바꿈
df_age = pd.Series(age_code)
df_age
```

```python
# 결측치 제거
df_group = df_group.dropna(axis=0)
df_group = df_group.dropna(axis=1)
df_group
```

```python
# df_age와 df_group 가로로 합침
pd.concat([df_age, df_group], axis=1).round(1)
```
